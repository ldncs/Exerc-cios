Está é uma base contendo dados descritivos de carros. O intuito desse Jupter Notebook é apenas treinar as minhas habilidades com a biblioteca Pandas. 

```python
# Importando biblioteca Pandas
import pandas as pd
```

```python
# Abrindo a base de dados
df = pd.read_csv("Automobile_data.csv")
```

```python
# Vizualizando o tamnanho da Base
df.shape
```

```python
# Verificando dados missing e os tipos de dados nas coluans
df.info()
```

```python
# Olhando as primeiras cinco linhas da base
df.head()
```

```python
# Verificando as últimas 5 linhas da base
df.tail()
```

```python
# Selecionando o carro mais caro da base
carro_maiscaro = df[['company','price']][df.price==df['price'].max()]

# Mostrando a linha com maior saláario
carro_maiscaro
```

```python
# Selecionando apenas os carros da companhia Toyota
toyota_carros = df[df.company == 'toyota']

# Mostrando apenas carros da companhia Toyota
toyota_carros
```

```python
# Contando o número total de carros por companhia
total_carros = df['company'].value_counts()

# Monstrando o total de carros por companhia
total_carros
```

```python
# Maior preço de cada carro por companhia
agrupando_companhias = df.groupby('company')[['price']].max()

# Resetando o índice e atualizando o DataFrame
agrupando_companhias_reset = agrupando_companhias.reset_index()

# Mostrando o maior preço de cada carro por companhia com índice resetado
agrupando_companhias_reset
```

```python
# Km média de cada carro por companhia
km_media_companhia = df.groupby('company')[['average-mileage']].mean()

# Resetando o index
km_media_companhia_reset = km_media_companhia.reset_index()

# Mostrando a km média de cada carro por companhia
km_media_companhia_reset
```

```python
# Classifique a base pela coluna de preços
classifique_carros_preço = df.sort_values(by = ['price', 'horsepower'], ascending = False)

# Mostrando a classificação de carros por preços
classifique_carros_preço
```

```python
# Criando dicionários para os carros alemães e japoneses
CarrosAlemaes = pd.DataFrame.from_dict({'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]})

CarrosJaponeses = pd.DataFrame.from_dict({'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]})

#criando um DataFrame a partir dos dois dicionários
nacionalidade_companhias = pd.concat([CarrosAlemaes, CarrosJaponeses], keys = ['Germany','Japan'])

# Mostrando o DataFrame
nacionalidade_companhias
```

```python
# DataFrames 
carros_preco = pd.DataFrame.from_dict({'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925 , 71400]})
carros_potencia = pd.DataFrame.from_dict({'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141, 80, 182 , 160]})

# Jutando os DataFrames
precos_potencia = pd.merge(carros_potencia, carros_preco, on = 'Company', how = 'inner')

# Mostrando o novo Dataframe
precos_potencia
```
